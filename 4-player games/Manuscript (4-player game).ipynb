```python
def run_4_player_prisoners_dilemma():

    # Empty the output file
    open("output.txt", "w").close()

    print ("\n---- 4 Player Prisoner's Dilemma ----")
    # Pure strategy Nash eq = (D,D,D,D)
    game_name = "4 by 4"
    players_names = ['player1','player2', 'player3', 'player4']
    players_strategies = {}
    players_strategies['player1'] = ['C','D']
    players_strategies['player2'] = ['C','D']
    players_strategies['player3'] = ['C','D']
    players_strategies['player4'] = ['C','D']
    payoff_matrix = {}
    payoff_matrix[(('player1','C'),('player2','C'),('player3','C'),('player4','C'))] = {'player1':3,'player2':7,'player3':2,'player4':6}
    payoff_matrix[(('player1','C'),('player2','C'),('player3','C'),('player4','D'))] = {'player1':1,'player2':9,'player3':4,'player4':8}
    payoff_matrix[(('player1','C'),('player2','C'),('player3','D'),('player4','C'))] = {'player1':7,'player2':2,'player3':3,'player4':9}
    payoff_matrix[(('player1','C'),('player2','C'),('player3','D'),('player4','D'))] = {'player1':4,'player2':8,'player3':6,'player4':1}

    payoff_matrix[(('player1','C'),('player2','D'),('player3','C'),('player4','C'))] = {'player1':9,'player2':3,'player3':7,'player4':2}
    payoff_matrix[(('player1','C'),('player2','D'),('player3','C'),('player4','D'))] = {'player1':2,'player2':6,'player3':1,'player4':7}
    payoff_matrix[(('player1','C'),('player2','D'),('player3','D'),('player4','C'))] = {'player1':5,'player2':9,'player3':8,'player4':3}
    payoff_matrix[(('player1','C'),('player2','D'),('player3','D'),('player4','D'))] = {'player1':6,'player2':1,'player3':2,'player4':8}

    payoff_matrix[(('player1','D'),('player2','C'),('player3','C'),('player4','C'))] = {'player1':8,'player2':4,'player3':9,'player4':3}
    payoff_matrix[(('player1','D'),('player2','C'),('player3','C'),('player4','D'))] = {'player1':3,'player2':7,'player3':2,'player4':6}
    payoff_matrix[(('player1','D'),('player2','C'),('player3','D'),('player4','C'))] = {'player1':6,'player2':8,'player3':7,'player4':4}
    payoff_matrix[(('player1','D'),('player2','C'),('player3','D'),('player4','D'))] = {'player1':9,'player2':2,'player3':3,'player4':7}

    payoff_matrix[(('player1','D'),('player2','D'),('player3','C'),('player4','C'))] = {'player1':4,'player2':9,'player3':5,'player4':8}
    payoff_matrix[(('player1','D'),('player2','D'),('player3','C'),('player4','D'))] = {'player1':7,'player2':3,'player3':6,'player4':1}
    payoff_matrix[(('player1','D'),('player2','D'),('player3','D'),('player4','C'))] = {'player1':8,'player2':5,'player3':4,'player4':9}
    payoff_matrix[(('player1','D'),('player2','D'),('player3','D'),('player4','D'))] = {'player1':9,'player2':9,'player3':9,'player4':9}

    # Define an instance of the game
    PD = game(game_name, players_names, players_strategies, payoff_matrix)
    # Define an instance of the NashEqFinder
    NashEqFinderInst = NashEqFinder(PD, stdout_msgs=False, stdout_timing=False)
    [Nash_equilibria, exit_flag, game_payoff_matrix] = NashEqFinderInst.optlangRun()
    print("The old equilibria are:", Nash_equilibria)
    # Nash_equilibria is a list of [[('player1', 'C'), ('player2', 'C'), ('player3', 'C'), ('player4', 'C')], [('player1', 'D'), ('player2', 'D'), ('player3', 'D'), ('player4', 'D')]
    # but we need it to be a list of [(('player1', 'C'), ('player2', 'C'), ('player3', 'C'), ('player4', 'C')), (('player1', 'D'), ('player2', 'D'), ('player3', 'D'), ('player4', 'D'))]
    original_nasheq_cells = [(cell[0], cell[1], cell[2], cell[3]) for cell in Nash_equilibria]
    NashEqFinderInst.newEquilibria(nasheq_cells=[(('player1','C'), ('player2','C'), ('player3','C'), ('player4','C'))], strategies=['C', 'D'], original_nasheq_cells=original_nasheq_cells)

    # Find the new equilibria
    [Nash_equilibria, exit_flag, game_payoff_matrix] = NashEqFinderInst.optlangRun()
    
```

```python
run_4_player_prisoners_dilemma()
```
